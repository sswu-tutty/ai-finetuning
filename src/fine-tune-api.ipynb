````python
import json
import re

# 데이터 정제 함수
def clean_text(text):
    # 불필요한 공백 및 특수 문자 제거
    text = re.sub(r'\s+', ' ', text).strip()
    text = re.sub(r'[^\w\s]', '', text)
    return text

# 파일 정제 및 변환 함수
def preprocess_and_convert(input_filepath, output_filepath, system_message):
    try:
        with open(input_filepath, 'r', encoding='utf-8') as infile, open(output_filepath, 'w', encoding='utf-8') as outfile:
            data = json.load(infile)
            print(f"데이터 로드 성공: {len(data['data'])} 개 항목 처리")

            # 데이터를 순회하며 summary_entire와 summary_section에서 원문 및 요약을 처리
            for item in data['data']:
                # summary_entire 처리
                if 'summary_entire' in item:
                    for summary_item in item['summary_entire']:
                        if 'orginal_text' in summary_item and 'summary_text' in summary_item:
                            cleaned_input = clean_text(summary_item['orginal_text'])
                            cleaned_output = clean_text(summary_item['summary_text'])

                            formatted_data = {
                                "messages": [
                                    {"role": "system", "content": system_message},
                                    {"role": "user", "content": cleaned_input},
                                    {"role": "assistant", "content": cleaned_output}
                                ]
                            }
                            # JSONL 형식으로 저장 (각 줄에 하나씩 저장)
                            json.dump(formatted_data, outfile, ensure_ascii=False)
                            outfile.write('\n')  # 각 JSON 객체를 새로운 줄에 저장
                            print(f"처리된 데이터: {cleaned_input[:30]}... -> {cleaned_output[:30]}...")

                # summary_section 처리
                if 'summary_section' in item:
                    for summary_item in item['summary_section']:
                        if 'orginal_text' in summary_item and 'summary_text' in summary_item:
                            cleaned_input = clean_text(summary_item['orginal_text'])
                            cleaned_output = clean_text(summary_item['summary_text'])

                            formatted_data = {
                                "messages": [
                                    {"role": "system", "content": system_message},
                                    {"role": "user", "content": cleaned_input},
                                    {"role": "assistant", "content": cleaned_output}
                                ]
                            }
                            # JSONL 형식으로 저장 (각 줄에 하나씩 저장)
                            json.dump(formatted_data, outfile, ensure_ascii=False)
                            outfile.write('\n')  # 각 JSON 객체를 새로운 줄에 저장
                            print(f"처리된 데이터: {cleaned_input[:30]}... -> {cleaned_output[:30]}...")

            print(f"정제된 데이터가 '{output_filepath}' 파일에 성공적으로 저장되었습니다.")
    except Exception as e:
        print(f"오류 발생: {e}")

# 경로 설정
input_filepath = '/content/drive/MyDrive/finetuning/openai-api-lab-main/dataset/summary.json'  # 실제 경로 확인
output_filepath = '/content/drive/MyDrive/finetuning/openai-api-lab-main/dataset/summary_for_finetuning.jsonl'  # JSONL 포맷으로 저장

system_message = "너는 긴 논문 텍스트를 입력받아 중요한 정보를 요약하여 제공하는 에이전트이다."

# 함수 실행
preprocess_and_convert(input_filepath, output_filepath, system_message)
````

# 파일 업로드

훈련데이터와 검증데이터 open api를 통해서 파일 업로드 하기

````python
import os

from google.colab import drive
drive.mount('/content/drive')


training_file_name = "/content/drive/MyDrive/finetuning/openai-api-lab-main/dataset/summary_for_finetuning_train.jsonl"
validation_file_name = "/content/drive/MyDrive/finetuning/openai-api-lab-main/dataset/summary_for_finetuning_validation.jsonl"

print(os.path.exists(training_file_name))  # True가 출력되어야 합니다.
print(os.path.exists(validation_file_name))  # True가 출력되어야 합니다.
````

````python
def upload_file(file_name: str, purpose: str) -> str:
    with open(file_name, "rb") as file_fd:
        response = openai.File.create(file=file_fd, purpose=purpose)
    return response.id

training_file_id = upload_file(training_file_name, "fine-tune")
validation_file_id = upload_file(validation_file_name, "fine-tune")

print("Training file ID:", training_file_id)
print("Validation file ID:", validation_file_id)
````

# 모델 미세 튜닝 job 생성

접미사 suffix="paper-summary"로 설정(논문 요약에 맞도록)

````python
import openai


# OpenAI API 키 설정
openai.api_key = userdata.get('OPENAI_API_KEY')

print("Training file ID:", training_file_id)
print("Validation file ID:", validation_file_id)

# Fine-tuning 작업 생성
response = openai.FineTuningJob.create(  # 'FineTune.create' 사용
    training_file=training_file_id,    # 업로드한 훈련 파일 ID
    validation_file=validation_file_id,  # 업로드한 검증 파일 ID
    model="gpt-4o-mini-2024-07-18",  # 모델 이름 (예: "gpt-4o-mini-2024-07-18")
    suffix="paper-summary",  # 파인튜닝 작업의 접미사
)

# 작업 ID 추출
job_id = response['id']

print("Job ID:", job_id)  # 작업 ID 출력
print("Status:", response['status'])  # 파인튜닝 작업 상태 출력
````

# API활용예시(프론트)

````python
import React, { useState } from 'react';

const SummarizePaper = () => {
  const [inputText, setInputText] = useState('');
  const [summary, setSummary] = useState('');
  const [loading, setLoading] = useState(false);

  // OpenAI API 호출 함수
  const getSummary = async () => {
    setLoading(true);

    try {
      const response = await fetch('https://api.openai.com/v1/chat/completions', {
        method: 'POST',
        headers: {
          'Content-Type': 'application/json',
          'Authorization': `Bearer ${process.env.REACT_APP_OPENAI_API_KEY}`,
        },
        body: JSON.stringify({
          model: "fine-tuned-model-id", // 여기서 fine-tuned 모델 ID를 사용
          messages: [
            { role: "system", content: "너는 긴 논문 텍스트를 입력받아 중요한 정보를 요약하여 제공하는 에이전트이다." },
            { role: "user", content: inputText }  // 사용자가 입력한 논문 텍스트
          ]
        })
      });

      const data = await response.json();
      setSummary(data.choices[0].message.content);  // 요약된 결과를 출력
    } catch (error) {
      console.error("Error fetching data: ", error);
      setSummary('An error occurred while fetching the summary.');
    }

    setLoading(false);
  };

  return (
    <div>
      <h1>논문 요약하기</h1>
      <textarea
        placeholder="논문 내용을 입력하세요"
        value={inputText}
        onChange={(e) => setInputText(e.target.value)}
        rows="10"
        cols="50"
      ></textarea>
      <br />
      <button onClick={getSummary} disabled={loading}>
        {loading ? '요약 중...' : '요약하기'}
      </button>
      {summary && <div><h2>요약된 내용:</h2><p>{summary}</p></div>}
    </div>
  );
};

export default SummarizePaper;
````
